```
# Question 2: For the fuel type coal, what is the percentage change in the fuel cost per unit burned in 1998 compared to 1994?

url ='https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel_data = pd.read_csv(url, error_bad_lines = False)
fuel_data.describe(include = 'all')
```

```
fuel_data.isnull().sum()
```

```
# Use groupby to count the number of each unique value in the fuel unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
```

```
fuel_data[['fuel_unit']] = fuel_data[['fuel_unit']].fillna(value = 'mcf')
```

```
# Check if missing values have been filled

fuel_data.isnull().sum()
```

```
fuel_data.describe()
```

```
fuel = fuel_data.groupby('report_year').count()
fuel
```

```
fuel.pct_change(periods=4)
```

The percentage change in the fuel cost per unit burned in 1998 compared to 1994 is -0.103644 or about -10.36%
